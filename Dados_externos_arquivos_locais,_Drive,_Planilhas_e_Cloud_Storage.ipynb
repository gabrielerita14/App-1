```
import pandas as pd
import numpy as np

# Dados fornecidos
dados = {
    'Data': ['15/01/2025', '15/01/2025', '15/01/2025', '15/01/2025', '15/01/2025'],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Porto Alegre', 'Salvador'],
    'Temperatura Máxima (°C)': [30.5, 35.0, 24.0, 28.0, 31.0],
    'Temperatura Mínima (°C)': [22.0, 25.0, 18.0, 20.0, 24.5],
    'Precipitação (mm)': [12.0, np.nan, 8.0, 15.0, np.nan],
    'Umidade Relativa (%)': [78, 70, np.nan, 82, 80]
}

# Criação do DataFrame
df = pd.DataFrame(dados)

# Substituição dos valores ausentes na coluna Precipitação pela média
df['Precipitação (mm)'].fillna(df['Precipitação (mm)'].mean(), inplace=True)

# Substituição dos valores ausentes na coluna Umidade Relativa pela mediana
df['Umidade Relativa (%)'].fillna(df['Umidade Relativa (%)'].median(), inplace=True)

# Adição da coluna Amplitude Térmica
df['Amplitude Térmica (°C)'] = df['Temperatura Máxima (°C)'] - df['Temperatura Mínima (°C)']

# Criação de um novo DataFrame com cidades com Temperatura Máxima acima de 30°C
df_acima_30 = df[df['Temperatura Máxima (°C)'] > 30]

# Reordenação das colunas
df = df[['Data', 'Cidade', 'Temperatura Máxima (°C)', 'Temperatura Mínima (°C)', 'Amplitude Térmica (°C)', 'Precipitação (mm)', 'Umidade Relativa (%)']]

print("DataFrame Original:\n", df)
print("\nDataFrame com Temperatura Máxima acima de 30°C:\n", df_acima_30)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dados fornecidos
dados = {
    'Data': ['15/01/2025', '15/01/2025', '15/01/2025', '15/01/2025', '15/01/2025'],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Porto Alegre', 'Salvador'],
    'Temperatura Máxima (°C)': [30.5, 35.0, 24.0, 28.0, 31.0],
    'Temperatura Mínima (°C)': [22.0, 25.0, 18.0, 20.0, 24.5],
    'Precipitação (mm)': [12.0, np.nan, 8.0, 15.0, np.nan],
    'Umidade Relativa (%)': [78, 70, np.nan, 82, 80]
}

# Criação do DataFrame
df = pd.DataFrame(dados)

# Substituição dos valores ausentes na coluna Precipitação pela média
df['Precipitação (mm)'].fillna(df['Precipitação (mm)'].mean(), inplace=True)

# Substituição dos valores ausentes na coluna Umidade Relativa pela mediana
df['Umidade Relativa (%)'].fillna(df['Umidade Relativa (%)'].median(), inplace=True)

# Adição da coluna Amplitude Térmica
df['Amplitude Térmica (°C)'] = df['Temperatura Máxima (°C)'] - df['Temperatura Mínima (°C)']

# Criação de um novo DataFrame com cidades com Temperatura Máxima acima de 30°C
df_acima_30 = df[df['Temperatura Máxima (°C)'] > 30]

# Reordenação das colunas
df = df[['Data', 'Cidade', 'Temperatura Máxima (°C)', 'Temperatura Mínima (°C)', 'Amplitude Térmica (°C)', 'Precipitação (mm)', 'Umidade Relativa (%)']]

print("DataFrame Original:\n", df)
print("\nDataFrame com Temperatura Máxima acima de 30°C:\n", df_acima_30)

# Dados fictícios para a evolução da temperatura
horas = list(range(25))
temperaturas = [15 + (1.25 * hora) if hora <= 12 else 30 - (0.5 * (hora - 12)) for hora in horas]

# Criação do gráfico de linha
plt.figure(figsize=(10, 5))
plt.plot(horas, temperaturas, marker='o', linestyle='-')
plt.title('Evolução da Temperatura Durante o Dia')
plt.xlabel('Hora do Dia')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.show()

# Dados fictícios de vendas
dados_vendas = {
    'Dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom'],
    'Vendas': [200, 220, 250, 210, 230, 300, 280],
    'Clientes': [20, 22, 25, 21, 23, 30, 28],
    'Lucro': [100, 110, 125, 105, 115, 150, 140]
}

# Criação do DataFrame
df_vendas = pd.DataFrame(dados_vendas)

# Gráfico de barras para total de vendas por dia
plt.figure(figsize=(10, 5))
sns.barplot(x='Dia', y='Vendas', data=df_vendas)
plt.title('Total de Vendas por Dia')
plt.xlabel('Dia da Semana')
plt.ylabel('Total de Vendas')
plt.show()

# Gráfico de dispersão para número de clientes x total de vendas
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Clientes', y='Vendas', data=df_vendas)
plt.title('Número de Clientes vs Total de Vendas')
plt.xlabel('Número de Clientes')
plt.ylabel('Total de Vendas')
plt.show()

# Heatmap para correlação entre variáveis
plt.figure(figsize=(10, 5))
# Select only numerical columns for correlation calculation
correlacao = df_vendas[['Vendas', 'Clientes', 'Lucro']].corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Correlação Entre Variáveis de Vendas')
plt.show()
```
